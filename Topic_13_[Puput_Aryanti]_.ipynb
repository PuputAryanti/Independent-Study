### MODEL 1

```python
#Buat model dengan menggunakan data train

from sklearn.linear_model import LinearRegression

linreg = LinearRegression()
linreg.fit(train_X, train_y)
```

```python
train_X2 = sm.add_constant(train_X)   #menambahkan variabel konstan

#mencari model regresi
model = sm.OLS(train_y, train_X2).fit()
model.summary()
```

Dengan syntax di atas, diperoleh output dengan rincian sebagai berikut:

1. Metode yang digunakan adalah least squares
2. Nilai R-square yang diperoleh yaitu 0,941 atau 94,1%
3. Uji simultan (Uji F) menggunakan statistik uji F
(menguji model secara simultan/ bersama-sama)
  * Hipotesis:
    
    H0: Koefisien regresi tidak signifikan
    
    H1: Minimal ada 1 koefisien regresi yang signifikan
  * Alpha = 0,05
  * Statistik Uji: stat-F
    
    pvalue = 0.00
  * Kriteria Uji: Tolak H0 jika pvalue < alpha
    
    Karena pvalue (0.00) < alpha (0.05) maka H0 ditolak
  * Kesimpulan: Dengan tingkat kepercayaan 95%, dapat disimpulkan bahwa terdapat minimal satu koefisien regresi yang signifikan
4. Uji partial (Uji t) menggunakan statistik uji t

  (karena pada uji F, diperoleh kesimpulan bahwa ada minimal satu koefisien regresi yang signifikan)
maka diperlukan uji partial untuk mengetahui variabel mana saja yang signifikan)
  * Hipotesis:
  
    H0: Variabel independent tidak berhubungan linear dengan variabel dependent
    
    H1: Variabel independent berhubungan linear dengan variabel dependent
  * Alpha = 0,05
  * Statistik Uji: stat-t
    
    pvalue lengkap terdapat pada tabel
  * Kriteria Uji: Tolak H0 jika pvalue < alpha
  * Kesimpulan: Dengan tingkat kepercayaan 95%, dapat disimpulkan bahwa ada 40 variabel yang memiliki nilai pvalue < 0,05 maka varaibel tersebut berhubungan linear dengan variabel dependent

```python
#Menampilkan coeffient dari model
print(f"intercept: {linreg.coef_}")
```

```python
#Menghitung r square score
r_sq = linreg.score(train_X, train_y)
print(f'coefficient of determination: {r_sq}')
```

#### Uji Asumsi Klasik

```python
#Residual berdistribusi normal
res1 = train_y-linreg.predict(train_X)

from scipy.stats import kstest

kstest(res1, 'norm')
```

1. Asumsi pertama yaitu residual berdistribusi normal

  * Hipotesis
    H0: Residual berdistribusi normal
    
    H1: Residual tidak berdistribusi normal
  * Alpha = 0,05
  * Statistik Uji:
    
    pvalue = 0.00
  * Kriteria Uji: Tolak H0 jika pvalue < alpha
    
    Karena pvalue (0.00) < alpha (0.05) maka H0 ditolak
  * Kesimpulan: Dengan tingkat kepercayaan 95%, dapat disimpulkan bahwa residual tidak berdistribusi normal

2. Asumsi kedua yaitu tidak adanya multikolonieritas antar variabel independent

  Pada output model summary diketahui bahwa nilai eigenvalue terkecil yaitu  8.22e-25 yang mengindikasikan bahwa terdapat multikolinearitas antara variabel independent

Selain kedua asumsi tersebut, masih terdapat asumsi lain yaitu tidak terdapat heteroskedastisitas (varians dari error tidak terpengaruh oleh faktor lain)
