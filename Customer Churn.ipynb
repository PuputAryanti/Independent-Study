Dari output di atas, terdapat variabel 'TotalCharges' dengan tipe data object, seharusnya variabel tersebut merupakan variabel numeric (integer/float). Selain itu pada variabel SeniorCitizen yang seharusnya merupakan data kategorik tetapi memiliki tipe data integer

```python
#Untuk lebih yakin mengenai tipe data, cek unique value di TotalCharges
pd.unique(df['TotalCharges'])
```

Dapat diketahui bahwa value dari 'TotalCharges' merupakan angka, maka kita bisa merubah tipe data variabel tersebut menjadi float menggunakan fungsi .astype(float) atau pd.to_numeric().

* df['TotalCharges'] = df.TotalCharges.astype(float)
* df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

jika syntax di atas dijalankan, maka akan ada error, ini dikarenakan object/str tidak bisa diubah ke float jika ada empty string, atau terdapat simbol/huruf.

Dalam variabel 'TotalCharges' terdapat empty string, maka gunakan tambahan parameter 'errors = coerce' yg artinya jika ada error, maka value tersebut menjadi null (NaN).

* errors = ignore  > jika ada error akan diabaikan
* errors = raise   > akan error/ tidak bisa diubah (default)
* errors = coerce  > jika ada error, value tersebut berubah menjadi NaN

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors="coerce")
```

```python
#Cek unique value di variabel SeniorCitizen
pd.unique(df['SeniorCitizen'])
```

```python
#Karena value dalam variabel SeniorCitizen hanya 0 dan 1 maka kita dapat mengubahnya menjadi data kategorik/object
df['SeniorCitizen']=df['SeniorCitizen'].astype('object')
```

```python
#Setelah tipe data diubah, cek kembali tipe data
df.info()
```

Tipe data 'TotalCharges' dan 'SeniorCitizen' sudah sesuai.

### 3. Outlier Handling

Dalam dataset ini terdapat tiga variabel numeric yaitu 'tenure', 'MonthlyCharges', dan 'TotalCharges'.
Outlier merupakan data yang terpencil/nilai ekstrem. Untuk melihat apakah terdapat outlier dalam suatu variabel dapat menggunakan boxplot atau perhitungan manual.

```python
sns.boxplot(df['tenure'])
```

```python
sns.boxplot(df['MonthlyCharges'])
```

```python
sns.boxplot(df['TotalCharges'])
```

Pada perhitungan manual, kita perlu menghitung nilai kuartil 1, kuartil 3, IQR (Jangkauan Interkuartil), lower bound (batas bawah), dan upper bound (batas atas) dari setiap variabel.

Nilai yang lebih dari batas atas atau kurang dari batas bawah akan dianggap sebagai outlier.

```python
#outlier tenure

Q1_T = df['tenure'].quantile(0.25)
Q3_T = df['tenure'].quantile(0.75)
IQR_T = Q3_T - Q1_T
lower_bound_T = Q1_T - 1.5*IQR_T
upper_bound_T = Q3_T + 1.5*IQR_T

outlier_T = (df[df['tenure'] > upper_bound_T]) | (df[df['tenure'] < lower_bound_T])
outlier_T
```

```python
#outlier MonthlyCharges

Q1_MC = df['MonthlyCharges'].quantile(0.25)
Q3_MC = df['MonthlyCharges'].quantile(0.75)
IQR_MC = Q3_MC - Q1_MC
lower_bound_MC = Q1_MC - 1.5*IQR_MC
upper_bound_MC = Q3_MC + 1.5*IQR_MC

outlier_MC = (df[df['MonthlyCharges'] > upper_bound_MC]) | (df[df['MonthlyCharges'] < lower_bound_MC])
outlier_MC
```

```python
#outlier TotalCharges

Q1_TC = df['TotalCharges'].quantile(0.25)
Q3_TC = df['TotalCharges'].quantile(0.75)
IQR_TC = Q3_TC - Q1_TC
lower_bound_TC = Q1_TC - 1.5*IQR_TC
upper_bound_TC = Q3_TC + 1.5*IQR_TC

outlier_TC = (df[df['TotalCharges'] > upper_bound_TC]) | (df[df['TotalCharges'] < lower_bound_TC])
outlier_TC
```

Dari boxplot maupun perhitungan manual dapat diketahui bahwa tidak ada value yang lebih dari batas atas atau yang kurang dari batas bawah, maka dapat disimpulkan tidak ada outlier pada variabel 'tenure', 'MonthlyCharges', dan 'TotalCharges'.

### Korelasi

Untuk menghitung korelasi, data harus berupa numerical data, maka data categorical harus diubah ke data biner (0 dan 1)/ variabel dummy

Nilai korelasi berkisar antara -1 sampai 1, dimana semakin mendekati nilai -1 atau 1 maka hubungan antar variabel semakin kuat
* 0: no correlation
* 1: positive correlation (jika satu variabel meningkat, maka variabel lainnya akan meningkat juga)
* -1: negative correlation (jika satu variabel meningkat, maka variabel lainnya akan menurun)

Fungsi yang digunakan untuk mengitung korelasi antar variabel dalam dataframe adalah corr()

Visualisasi yang bisa digunakan yaitu scatterplot dan heatmap

```python
df2 = pd.get_dummies(df, drop_first = True)
df2.head()
```

```python
df2.corr()
```

Variabel tenure dan TotalCharges memiliki nilai korelasi sebesar 0,82588. Nilai korelasi ini bertanda positif yang artinya jika tenure meningkat maka TotalCharges juga akan maningkat.

```python
#Visualisasikan nilai korelasi di atas dengan heatmap
sns.heatmap(df2.corr(),cmap="Blues")
```

Semakin gelap warna dalam heatmap maka korelasi tersebut semakin kuat dan bertanda positif, jiak warna semakin terang artinya korelasi tersebut semakin kuat dan bertanda negatif.

```python
#Kita juga bisa menghitung korelasi variabel lain dengan variabel churn
df3=df2.corr()['Churn_Yes'].sort_values().reset_index()
df3
```

Dari output di atas, terlihat tidak ada nilai korelasi di atas 4 atau di bawah -4 yang artinya korelasi variabel lain dengan variabel churn lemah.

```python
#Untuk memudahkan  dalam membaca nilai korelasi, kita bisa menvisualisasikannya ke dalam barplot
df3=df3[~df3['index'].str.contains('Churn_Yes')]

chart = sns.barplot(data=df3,x="index",y="Churn_Yes")

for item in chart.get_xticklabels():
  item.set_rotation(90)
```

Dari barplot di atas, dapat diketahui bahwa sebagian besar variabel berkorelasi lemah dengan variabel churn, hal ini dilihat dari nilai korelasi yang dibawah 4 atau di atas -4.

Korelasi tertinggi adalah antara variabel InternetService_Fiber optic dengan variabel churn di mana nilai korelasi sebesar 0.307.
